## CSV까지 완료

```python
from bs4 import BeautifulSoup as bs
import urllib.request as req  
import urllib.parse as prs
import time
import gc
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
```

```python
url = 'http://www.ssg.com/search.ssg?target=all&query=+%EC%9D%B8%ED%85%8C%EC%9D%B4%ED%81%AC&srch_area=sssg&page='
```

```python
def url_get(url):
    HTML = req.urlopen(url)
    page = bs(HTML.read(), 'html.parser')
    infotable = page.find('ul',{'class':'cunit_thmb_lst'}).findAll('li')
    target_list = list()
    for info in infotable:
        _ = info.find('div',{'class':'title'})
        if (str(type(_.strong)) == "<class 'NoneType'>" ) : continue
        if ((_.strong.get_text() == '닥터넛츠') or (_.strong.get_text() == '인테이크')):
            target_list.append([_.strong.get_text(), #회사브랜드
                                _.a['href'],         #상품주소
                                _.a.get_text()])     #상품명
    return target_list    
```

```python
url_list = list()
for n in [1,2]:
    url_list+=url_get(url+str(n))
print("총 가져온 URL 수: ",len(url_list))
```

```python
driver = webdriver.Chrome('chromedriver2.exe',chrome_options=options)
bucket = list()
for url in url_list:
    gc.collect()
    print(url[2])
    print('http://www.ssg.com'+url[1])
    driver.get('http://www.ssg.com'+url[1])
    driver.implicitly_wait(1)
    
    click_index = 1
    flag = 11
    while True:
        HTML = driver.page_source
        page = bs(HTML, 'html.parser')
        if str(type(page.find('tbody',{'id':'cdtl_cmt_tbody'}))) == "<class 'NoneType'>": print(type(page.find('tbody',{'id':'cdtl_cmt_tbody'}))); break
        cmt_list = page.find('tbody',{'id':'cdtl_cmt_tbody'}).findAll('tr')[::2]
        for _ in cmt_list:
            usr = _.find('td', {'class':'user'}).get_text()
            dte = _.find('td', {'class':'date'}).get_text().strip()
            rte = _.find('span', {'class':"sp_cdtl cdtl_cmt_per"})['style'].split(':')[1]
            cmt = _.find('div', {'class':'cdtl_cmt_tx'}).get_text().strip()
            itm = url[2]
            if str(type(_.find('div',{'class':'cdtl_cmt_tx2'}))) != "<class 'NoneType'>":
                opt = _.find('div',{'class':'cdtl_cmt_tx2'}).get_text()
            else : opt = None
            bucket.append([usr, dte, rte, cmt, itm, opt])
            
        try: 
            driver.find_element_by_xpath('//*[@id="comment_navi_area"]/a['+str(click_index)+']').click()
            time.sleep(1)
            click_index = click_index + 1
            if click_index == flag: 
                click_index = 3
                flag = 13
        except : break
  
```
